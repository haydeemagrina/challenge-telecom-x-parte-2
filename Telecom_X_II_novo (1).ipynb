```python
# Imports
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import xgboost as xgb

df = pd.read_csv('/content/dados_tratados_tudo.csv')

# Remover colunas irrelevantes
colunas_remover = ['customerID', 'Custo Diário', 'Faixa Custo Diário']
df = df.drop(columns=colunas_remover)

# Transformar a variável alvo em numérica
df['Churn'] = df['Churn'].astype(int)

# Separar variáveis numéricas e categóricas
cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include=np.number).columns.drop('Churn')

# One-hot encoding para variáveis categóricas
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)

# Separar features e target
X = df_encoded.drop('Churn', axis=1)
y = df_encoded['Churn']

# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Criar o modelo XGBoost
model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
model.fit(X_train, y_train)

# Previsões
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Avaliação
acc = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)
cm = confusion_matrix(y_test, y_pred)

print(f"Acurácia: {acc:.4f}")
print(f"AUC-ROC: {roc_auc:.4f}")
print("Matriz de Confusão:")
print(cm)

# Curva ROC
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, label=f'XGBoost (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Curva ROC')
plt.legend()
plt.show()

# Importância das variáveis
xgb.plot_importance(model, max_num_features=10)
plt.title('Importância das Variáveis')
plt.show()
```

```python
print(f'Acurácia Dummy: {dummy.score(x_teste, y_teste)}')
print(f'Acurácia Árvore de Decisão: {arvore.score(x_teste_numeric, y_teste)}')
print(f'Acurácia KNN: {knn.score(x_teste_numeric_normalizado, y_teste)}')
print(f'Acurácia XGBoost: {acc}')
```

# Avaliação da AUC-ROC de todos os modelos (Dummy, Árvore de Decisão, KNN e XGBoost)

```python
# Dummy Classifier
dummy = DummyClassifier(strategy='most_frequent')
dummy.fit(X_train, y_train)
y_pred_dummy = dummy.predict(X_test)
y_proba_dummy = dummy.predict_proba(X_test)[:, 1]
acc_dummy = accuracy_score(y_test, y_pred_dummy)
roc_dummy = roc_auc_score(y_test, y_proba_dummy)

# Decision Tree
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)
y_proba_tree = tree.predict_proba(X_test)[:, 1]
acc_tree = accuracy_score(y_test, y_pred_tree)
roc_tree = roc_auc_score(y_test, y_proba_tree)

# KNN
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
y_proba_knn = knn.predict_proba(X_test)[:, 1]
acc_knn = accuracy_score(y_test, y_pred_knn)
roc_knn = roc_auc_score(y_test, y_proba_knn)

# XGBoost
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
y_proba_xgb = xgb_model.predict_proba(X_test)[:, 1]
acc_xgb = accuracy_score(y_test, y_pred_xgb)
roc_xgb = roc_auc_score(y_test, y_proba_xgb)

# Imprimir resultados
print(f"Acurácia Dummy: {acc_dummy:.4f}, AUC-ROC: {roc_dummy:.4f}")
print(f"Acurácia Árvore de Decisão: {acc_tree:.4f}, AUC-ROC: {roc_tree:.4f}")
print(f"Acurácia KNN: {acc_knn:.4f}, AUC-ROC: {roc_knn:.4f}")
print(f"Acurácia XGBoost: {acc_xgb:.4f}, AUC-ROC: {roc_xgb:.4f}")

# Curvas ROC
fpr_dummy, tpr_dummy, _ = roc_curve(y_test, y_proba_dummy)
fpr_tree, tpr_tree, _ = roc_curve(y_test, y_proba_tree)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_proba_knn)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_proba_xgb)

plt.figure(figsize=(8, 6))
plt.plot(fpr_dummy, tpr_dummy, label=f'Dummy (AUC = {roc_dummy:.2f})')
plt.plot(fpr_tree, tpr_tree, label=f'Tree (AUC = {roc_tree:.2f})')
plt.plot(fpr_knn, tpr_knn, label=f'KNN (AUC = {roc_knn:.2f})')
plt.plot(fpr_xgb, tpr_xgb, label=f'XGBoost (AUC = {roc_xgb:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Curva ROC Comparativa')
plt.legend()
plt.show()

# Importância das variáveis - XGBoost
xgb.plot_importance(xgb_model, max_num_features=10)
plt.title('Importância das Variáveis - XGBoost')
plt.show()
```

Análise por Modelo (com AUC):
Modelo	AUC-ROC	Interpretação
Dummy	0.50	Péssimo — desempenho igual a um chute aleatório
Tree	0.63	Fraco — alguma capacidade de separação
KNN	0.72	Moderado — já distingue bem churn vs não churn
XGBoost	0.78	Melhor — bom equilíbrio entre TPR e FPR

Conclusão:
XGBoost tem a melhor curva ROC e maior AUC (0.78), indicando que é o modelo mais confiável para prever churn.

KNN também é razoável, mas XGBoost supera em desempenho geral.

Árvore de Decisão é significativamente inferior em termos de AUC, apesar da boa acurácia.

# Ajuste fino (tuning) no XGBoost para tentar aumentar ainda mais o AUC

```python
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV

```

```python
pip show scikit-learn
```

```python
# Ajuste fino XGBoost - GridSearchCV
param_grid = {
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2]
}
grid_search = GridSearchCV(estimator=xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
                            param_grid=param_grid,
                            scoring='roc_auc',
                            cv=3, verbose=1)
grid_search.fit(X_train, y_train)
print("Melhores parâmetros GridSearch:", grid_search.best_params_)
print("Melhor AUC-ROC GridSearch:", grid_search.best_score_)

# Ajuste fino XGBoost - RandomizedSearchCV
param_dist = {
    'max_depth': [3, 4, 5, 6, 7, 8],
    'n_estimators': [50, 100, 150, 200],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}
random_search = RandomizedSearchCV(estimator=xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
                                   param_distributions=param_dist,
                                   n_iter=20,
                                   scoring='roc_auc',
                                   cv=3, verbose=1, random_state=42)
random_search.fit(X_train, y_train)
print("Melhores parâmetros RandomizedSearch:", random_search.best_params_)
print("Melhor AUC-ROC RandomizedSearch:", random_search.best_score_)
```

# Avaliação real com melhores parâmetros RandomizedSearch

```python
best_model = random_search.best_estimator_
y_pred_best = best_model.predict(X_test)
y_proba_best = best_model.predict_proba(X_test)[:, 1]
acc_best = accuracy_score(y_test, y_pred_best)
roc_best = roc_auc_score(y_test, y_proba_best)

print(f"Acurácia XGBoost Ajustado: {acc_best:.4f}, AUC-ROC: {roc_best:.4f}")

fpr_best, tpr_best, _ = roc_curve(y_test, y_proba_best)
plt.figure(figsize=(8, 6))
plt.plot(fpr_best, tpr_best, label=f'XGBoost Ajustado (AUC = {roc_best:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Curva ROC - ')
plt.legend()
plt.show()
```

Análise da Curva ROC:
Elemento	Descrição
TPR (True Positive Rate)	Mostra a taxa de acertos positivos reais do modelo. Quanto mais perto de 1, melhor.
FPR (False Positive Rate)	Mostra a taxa de falsos positivos. Quanto mais perto de 0, melhor.
Curva azul	Representa o desempenho do modelo.
Linha preta tracejada	Linha de referência (classificação aleatória, AUC = 0.5).
AUC = 0.81	Bom desempenho. Modelos com AUC entre 0.8 e 0.9 são considerados bons.

Interpretação:
O modelo XGBoost ajustado tem boa capacidade discriminativa para classificar corretamente entre as classes (ex: churn vs não churn).

O fato da curva estar bem acima da linha aleatória indica que o modelo aprendeu padrões úteis nos dados.

AUC de 0.81 sugere que, ao comparar aleatoriamente um cliente churn e um não churn, o modelo tem 81% de chance de rankear o churn corretamente acima do não churn.
